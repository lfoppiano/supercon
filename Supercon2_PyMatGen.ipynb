```python
%pip install pymatgen
from pymatgen.core.composition import Composition
import pandas as pd
import numpy as np
import re

! git clone https://github.com/lfoppiano/supercon.git
```

```python
#change to your directory
supercon2 = pd.read_csv("/content/supercon/data/supercon2_v23.04.06.csv")
display(supercon2)
```

```python
# Make a dataframe of formulas and critical temperature (with some other inportant info)

df1 = supercon2[['formatted_formula','criticalTemperature', 'appliedPressure', 'criticalTemperatureMeasurementMethod','doping', 'shape', 'materialClass', 'fabrication', 'substrate', 'title', 'doi']]
new_df = df1.rename(columns={'criticalTemperature': 'target'})
new_df = new_df.rename(columns={'formatted_formula': 'formula'})
display(new_df)
```

```python
#remove NaNs

no_nan = new_df.dropna(subset=['formula', 'target'])
display(no_nan)
```

```python
# Reduce formulas and remove formulas that can't be reduced

# create an empty dataframe to append reduced formulas and targets to
new_df = pd.DataFrame(columns=["formula", "target", "appliedPressure", 'criticalTemperatureMeasurementMethod','doping', 'shape', 'materialClass', 'fabrication', 'substrate', 'title', 'doi'])

# iterate over rows in the original dataframe
for index, row in no_nan.iterrows():
    try:
        # try to reduce the formula using pymatgen's Composition class
        composition = Composition(row["formula"])
        reduced_formula = composition.get_reduced_formula_and_factor()[0]
        # append the reduced formula and target to the new dataframe
        new_row = {"formula": reduced_formula, "target": row["target"], "appliedPressure" : row["appliedPressure"], 'criticalTemperatureMeasurementMethod':row['criticalTemperatureMeasurementMethod'],'doping':row['doping'], 'shape':row['shape'], 'materialClass':row['materialClass'], 'fabrication':row['fabrication'], 'substrate':row['substrate'], 'title':row['title'], 'doi':row['doi']}
        # new_df = new_df.append(new_row, ignore_index=True)
        new_df = pd.concat([new_df, pd.DataFrame(new_row, index=[0])], ignore_index=True)
    except:
        # if the formula can't be reduced, print a message and move on to the next row
        print(f"Formula {row['formula']} could not be reduced.")
        continue

display(new_df)
```

```python
# Clean up target column to have only floats
def extract_float(string):
    match = re.search(r'\d+\.?\d*', string)
    if match:
        return float(match.group())
    else:
        return None

# Apply the function to "target" column and create a new column "target_clean"
new_df['target_clean'] = new_df['target'].apply(extract_float)

# Replace the original "target" column with the cleaned values
new_df = new_df.drop(columns={'target'})
new_df = new_df.rename(columns={'target_clean':'target'})

# Remove rows with missing values and NaNs and bring target column to the front of the dataframe
new_df = new_df.dropna(subset=['formula', 'target'])
last_col = new_df.pop('target')
new_df.insert(1, 'target', last_col)

# Display the cleaned DataFrame
display(new_df)
```

```python
sorted = new_df.sort_values(by="target", ascending=False)
display(sorted)
```

```python
#further curation if you want to sort based on critical temperature

# Create a boolean mask for the rows where "target" is in the range of 0-300
mask = (sorted['target'] >= 0) & (sorted['target'] <= 300)

# Create a new DataFrame containing only the rows where "target" is in the range of 0-300
upto300K = sorted.loc[mask]

# Print the new DataFrame
display(upto300K)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Create a histogram of the "target" column
upto300K['target'].hist()

# Add labels and title
plt.xlabel('Critical Temperature (K)')
plt.ylabel('Frequency')
plt.title('Histogram of Critical Temperatures in SuperCon2 dataset')

# Show the plot
plt.show()
```
